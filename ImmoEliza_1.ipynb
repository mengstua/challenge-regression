```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import time

# importing the file 

df_RF= pd.read_csv('cleaned_file.csv')
X  = df_RF.iloc[:, :-1]
y = df_RF.iloc[:, -1]

feature_names = df_RF.columns

# set number of iteration 
iterations = 10
best_MSE = float('inf')
best_model = None
best_r2 = None
best_params = None
start_time = time.time()

feature_names = X.columns

for i in range(1, iterations +1 ):
    iteration_start = time.time()

    random_start = np.random.randint(0 , 10000)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = random_start, shuffle = True)

    n_stimators = np.random.randint(50,1000)
    max_depth = np.random.choice([None, 5, 10, 20, 30])

    # preparing data for training and testing

    model_RF = RandomForestRegressor(n_estimators=n_stimators, max_depth=max_depth, random_state=random_start)
    model_RF.fit(X_train, y_train)
        
    # Predicting the model

    Y_pred_RF = model_RF.predict(X_test)

    MSE = mean_squared_error(y_test, Y_pred_RF)
    r2 = r2_score(y_test, Y_pred_RF)

    if MSE < best_MSE:
        best_MSE = MSE
        best_r2 = r2
        best_model = model_RF
        best_params = {'n_estimators' : n_stimators, 'max_depth' : max_depth, 'random_stat' : random_start}


    iteration_time = time.time() - iteration_start
    print(f"Iteration {i}/{iterations} completed in {iteration_time:.2f} seconds - MSE: {mse:.4f}, R2: {r2:.4f}")

total_time = time.time() - start_time


# Feature importances from best RandomForest
importances = best_model.feature_importances_
importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

print("\n==== Summary ====")

print(f"Best MSE: {best_MSE:.4f}")
print(f"Best R2 Score: {best_r2:.4f}")
print(f"Best parameters: {best_params}")
print(f"Total time taken: {total_time:.2f} seconds")

print("\n==== Feature Importances (RandomForest, descending order) ====")
print(importance_df.to_string(index=False))

```
